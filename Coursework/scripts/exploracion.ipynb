```python
colums_freq = list(filter(lambda x: 'word_freq' in x, list(df_train.drop(['email_id','spam'], axis=1).columns)))
colums_no_freq = list(filter(lambda x: 'word_freq' not in x, list(df_train.drop(['email_id','spam'], axis=1).columns)))
```

```python
sns.heatmap(df_train.drop(['email_id','spam'],axis=1)[colums_no_freq].corr())
```

# Modelo de juguete

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.model_selection import train_test_split 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LogisticRegression
```

```python
df_train2 = df_train.drop('email_id', axis=1)
x_train, x_val, y_train, y_val = train_test_split(df_train2.drop('spam',axis=1), df_train2[['spam']], test_size=0.3, random_state=12) 
```

```python
len(colums_freq)
```

```python
svd = TruncatedSVD(n_components=48, n_iter=7, random_state=42)
svd.fit(x_train)
```

```python
np.cumsum(svd.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(svd.explained_variance_ratio_))
```

```python
ct = ColumnTransformer([('svd',TruncatedSVD(n_components=20, n_iter=7, random_state=42),colums_freq),
                        ('norm',MinMaxScaler(),colums_no_freq)])
```

```python
clf = Pipeline(
    steps=[("preprocessor", ct), ("classifier", LogisticRegression())]
)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_val)
y_pred_prob=clf.predict_proba(x_val)
```

```python
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(y_val, y_pred_prob[:,1]) #Lo hace segun la probabilidad de que sea 1 
```

```python
clf["classifier"].coef_
```
